```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read data from csv file
df = pd.read_csv('CFB2019.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#divide "Win_Loss" column into two different "Win" and "Loss" columns
dfn = df['Win-Loss'].str.split('-', expand = True)
```

```python
dfn
```

```python
#concatenate 'Win' and 'Loss' columns to Data Fram 
df['Win'] = dfn[0]
df['Loss'] = dfn[1]
df.head()
```

```python
#Drop "Win-Loss" column 
df.drop(columns = ['Win-Loss'], inplace=True)
```

```python
df
```

```python
team = df['Team']
team.head()
```

```python
#Convert 'Win' and 'Loss' string columns to numeric columns
df[['Win']] = df[['Win']].apply(pd.to_numeric , errors = 'coerce')
df[['Loss']] = df[['Loss']].apply(pd.to_numeric, errors = 'coerce')
```

```python
df_corr = df.corr()
```

```python
df_corr['Win']
```

```python
df_corr['Loss']
```

```python
#Correlation between 'Loss' and 'Off Plays' columns 
df_corr['Loss']['Off Plays']
```

```python
#Pass 'spearman' correlation method as a parameter
df_corr = df.corr(method = 'spearman')
```

```python
#Display  the only columns that the 'Win' column are bigger than %70
df_corr[df_corr['Win'] > 0.7]['Win']
```

```python
#Graph the data
plt.figure(figsize = (20,12))
sns.heatmap(df_corr, cmap='YlGnBu', annot=True)
```

```python
#Choose some columns instead of all columns to load less data                 
df_n = df_corr[df_corr['Win'] > 0.7][['Games', 'Off Plays', 'Off Yards', 'Off TDs', 'First Downs', 'Redzone Attempts', 'Win']]
plt.figure(figsize=(20,12))
sns.heatmap(df_n, cmap='YlGnBu', annot=True)
```
